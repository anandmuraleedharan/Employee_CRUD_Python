```python
class Employee:
    def __init__(self, name, address, vehicle):
        self.__name = name
        self.__address = address
        self.__vehicle = vehicle

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_address(self):
        return self.__address

    def set_address(self, address):
        self.__address = address

    def get_vehicle(self):
        return self.__vehicle

    def set_vehicle(self, vehicle):
        self.__vehicle = vehicle
        
    def __str__(self):
        string = "\nEmployee Name: " + self.__name + "; Employee Address: " + self.__address
        string += "\nVehicle Make: " + self.__vehicle.get_make() + "; Vehicle Model: " + self.__vehicle.get_model() + "; Vehicle Year: " + self.__vehicle.get_year().__str__() + "; Vehicle Mileage: " + self.__vehicle.get_mileage().__str__()
        return string
```

```python
class FullTimeEmployee(Employee):
    def __init__(self, name, address, vehicle, salary):
        super().__init__(name, address, vehicle)
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        self.__salary = salary

    def compute_compensation(self):
        salary = self.__salary
        if(salary <= 45000):
              compensation = (1 - 0.18) * salary
        elif(salary <= 82000):
              compensation = ((1 - 0.18) * 45000) + (1-0.28) * (salary - 45000)
        else:
              compensation = ((1 - 0.18) * 45000) + ((1-0.28) * (salary - 45000)) + ((1-0.33) * (salary - 82000))
        return compensation

    def __str__(self):
        string = super().__str__()
        string +=  "\nSalary: " + self.__salary.__str__() + "\n"
        return string
```

```python
class HourlyEmployee(Employee):
    def __init__(self, name, address, vehicle, hoursWorked, hourlyRate):
        super().__init__(name, address, vehicle)
        self.__hoursWorked = hoursWorked
        self.__hourlyRate = hourlyRate

    def get_hoursWorked(self):
        return self.__hoursWorked

    def set_hoursWorked(self, hoursWorked):
        self.__hoursWorked = hoursWorked

    def get_hourlyRate(self):
        return self.__hourlyRate

    def set_hourlyRate(self, hourlyRate):
        self.__hourlyRate = hourlyRate

    def compute_compensation(self):
        hoursWorked = self.__hoursWorked
        hourlyRate = self.__hourlyRate
        if(hoursWorked > 40):
            compensation = (40 * hourlyRate) + (40 * 1.8 * (hoursWorked - 40))
        else:
            compensation = 40 * hourlyRate
        return compensation

    def __str__(self):
        string = super().__str__()
        string +=  "\nHours Worked: " + self.__hoursWorked.__str__() + "; Hourly Rate: " + self.__hourlyRate.__str__() + "\n"
        return string
```

```python
class Consultant(Employee):
    def __init__(self, name, address, vehicle, hoursWorked, projectType):
        super().__init__(name, address, vehicle)
        self.__hoursWorked = hoursWorked
        self.__projectType = projectType

    def get_hoursWorked(self):
        return self.__hoursWorked

    def set_hoursWorked(self, hoursWorked):
        self.__hoursWorked = hoursWorked

    def get_projectType(self):
        return self.__projectType

    def set_projectType(self, projectType):
        self.__projectType = projectType

    def compute_compensation(self):
        hoursWorked = self.__hoursWorked
        projectType = self.__projectType
        if(projectType == 1):
              compensation = 55 * hoursWorked
        elif(projectType == 2):
              compensation = 55 * hoursWorked
        else:
              compensation = 85 * hoursWorked
        return compensation

    def __str__(self):
        string = super().__str__()
        string +=  "\nHours Worked" + self.__hoursWorked.__str__() + "; Project Type" + self.__projectType.__str__() + ";"
        return string
```

```python
class Vehicle:
    def __init__(self, make, model, year, mileage):
        self.__make = make
        self.__model = model
        self.__year = year
        self.__mileage = mileage

    def get_make(self):
        return self.__make

    def set_make(self, make):
        self.__make = make

    def get_model(self):
        return self.__model

    def set_model(self, model):
        self.__model = model

    def get_year(self):
        return self.__year

    def set_year(self, year):
        self.__year = year

    def get_mileage(self):
        return self.__mileage

    def set_mileage(self, mileage):
        self.__mileage = mileage
        
    def get_str(self):
        string = "Vehicle Make: " + self.__make + "; Vehicle Model: " + self.__model + "; Vehicle Year: " + str(self.__year)
        string += "; Vehicle Mileage: " + str(self.__mileage)
        return string
```

```python
# Driver program begins here
import pickle
empList = list()

def main():
    readFileData()
    acceptUserInput()

def acceptUserInput():
    printmenu()
    userInput = acceptValidIntInput("Your selection: ", [1,2,3,4,5])
    processOptions(userInput)

def processOptions(userInput):
    if(userInput == 1):
        processOption1()
    elif(userInput == 2):
        if(checkForEmptyFile()):
            print("The file is empty. Please add Employee Information")
            acceptUserInput()
        else:
            processOption2()
    elif(userInput == 3):
        if(checkForEmptyFile()):
            print("\nThe file is empty. Please add Employee Information")
            acceptUserInput()
        else:
            processOption3()
    elif(userInput == 4):
        if(checkForEmptyFile()):
            print("\nThe file is empty. Please add Employee Information")
            acceptUserInput()
        else:
            processOption4()
    else:
        processOption5()

def processOption1():
    empType = acceptValidIntInput("Enter Employee Type [1-Fulltime, 2-Hourly, 3-Consultant]: ", [1,2,3])

    name = input("Enter the name of the employee: ")
    address = input("Enter the address of the employee: ")
    make = input("Enter the make of the vehicle: ")
    model = input("Enter the model of the vehicle: ")
    year = acceptValidIntInput("Enter the year of manufacture: ")
    mileage = acceptValidIntInput("Enter the mileage: ")
    vehicle = Vehicle(make,model,year,mileage)

    if(empType==1):
        salary = acceptValidIntInput("Enter the salary: ")
        Obj = FullTimeEmployee(name,address,vehicle,salary)
    elif(empType==2):
        hoursWorked = acceptValidIntInput("Enter the hours worked: ")
        hourlyRate = acceptValidIntInput("Enter the hourly rate: ")
        Obj = HourlyEmployee(name,address,vehicle,hoursWorked,hourlyRate)
    else:
        hoursWorked = acceptValidIntInput("Enter the hours worked: ")
        projectType = acceptValidIntInput("Enter the project type: ",[1,2,3])
        Obj = Consultant(name,address,vehicle,hoursWorked,projectType)
    empList.append(Obj)
    acceptUserInput()

def processOption2():
    print("\nDetails of Full-Time Employees")
    for emp in empList:
        if(type(emp).__name__ == 'FullTimeEmployee'):
              print(emp)
    print("Details of Hourly Employees")
    for emp in empList:
        if(type(emp).__name__ == 'HourlyEmployee'):
              print(emp)
    print("Details of Consultants")
    for emp in empList:
        if(type(emp).__name__ == 'Consultant'):
              print(emp)
    acceptUserInput()

def processOption3():
    print("Employee Name and Compensation of all employees:")
    for emp in empList:
        print(emp.get_name() + ";" + str(emp.compute_compensation()) + ";")
    acceptUserInput()

def processOption4():
    print("Employees with Vehicle Mileage > 78000")
    for emp in empList:
        if(emp.get_vehicle().get_mileage() > 78000):
            vehicle_str = emp.get_vehicle().get_str()
            print("Employee Name:" + emp.get_name() + " ;" + vehicle_str)
    acceptUserInput()

def processOption5():
    confirmExit = input("Do you want to exit [Press Y or N]: ")
    if(confirmExit.casefold() != 'y'):
        acceptUserInput()
    else:
        saveObjToFile()
        exit()

def printmenu():
    print("=======Menu=======")
    print("1. To add an employee")
    print("2. To print the name and address of all employees")
    print("3. To print the name and compensation of all employees")
    print("4. To print the employee name with vehicle information for high mileage vehicles ")
    print("5. To exit program")

def acceptValidIntInput(str, list=None):
    try:
        entry = int(input(str))
        if(list):
            if(entry not in list):
                raise Exception("Invalid Option")
        return entry
    except:
        print("Please enter a valid option")
        return acceptValidIntInput(str,list)

def saveObjToFile():
    fb = open('empdata.dat', 'wb')
    for obj in empList:
        pickle.dump(obj, fb)
    fb.close()

def readFileData():
    try:
        fb = open('empdata.dat', 'rb')
    except FileNotFoundError:
        print("No File found; Add Employee data first")
        return
    else:
        try:
            while (True): 
                empList.append(pickle.load(fb))
        except EOFError: 
            fb.close()    

def checkForEmptyFile():
    if(len(empList) == 0):
        return True
    else:
        return False

main()
```
